```python
def diccionario_inverso(diccionario): 
#función que invierte un diccionario, detecta valores repetidos y agrupa sus claves asociadas, devolviendo el diccionario invertido y avisando de las repeticiones.     
    
    try:
        #Validación 1: que el diccionario no esté vacío
        if diccionario is None:
            raise ValueError ("El diccionario no puede ser None.")
            
        # Validación 2: que realmente sea un diccionario
        if not isinstance(diccionario, dict):
            raise TypeError ("El argumento debe ser un diccionario.")
    
    except (ValueError, TypeError):
        return None 
    


    diccionario_reves = {}
    valores_repetidos = {}
    
    for key, value in diccionario.items():
        
        if value in diccionario_reves:       
            if value not in valores_repetidos: 
                valores_repetidos[value] = [diccionario_reves[value]]
            valores_repetidos[value].append(key) 
            diccionario_reves[value]= valores_repetidos[value]
            
        
        else: 
            diccionario_reves.update({value:key})
        
    for value, key in valores_repetidos.items():
        print(f"la clave {value} está repetida, aqui tienes la lista de los valores asociados a esa key", valores_repetidos[value])
    
    return  f"el diccionario invertido es : {diccionario_reves}"

diccionario_solicitado= { "brand": "Ford","model": "Mustang","year": 1964, "category": "Ford", "marca":"Ford", "año": 1964, "marque":"Ford"}
print(diccionario_inverso (diccionario_solicitado))

```

## Ejercicio 2 
El cliente recibe una lista de datos y necesita generar dos listas, la primera donde estarán todos los elementos que pudieron convertirse en flotantes y la otra donde están los elementos que no pudieron convertirse. Ejemplo de la lista que recibe el cliente:
[ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

```python
def conversion (lista_datos): 
#función que recorre una lista (posiblemente anidada) y convierte todos los elementos numéricos a float, separando los elementos que no se pueden convertir.
    
    datos_convertidos_float= []
    datos_no_convertidos= []

    for dato in lista_datos: 
        
        if isinstance(dato, (list, tuple)):
            subdatos_convertidos_float, subdatos_no_convertidos= conversion(dato)
            datos_convertidos_float.extend(subdatos_convertidos_float)
            datos_no_convertidos.extend(subdatos_no_convertidos)
        
        else: 
            try:  
                datos_convertidos_float.append(float(dato))
            except ValueError: 
                datos_no_convertidos.append(dato)
            except TypeError: 
                datos_no_convertidos.append(dato)

    return(datos_convertidos_float,datos_no_convertidos)   

lista= [ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]
print(conversion(lista))

```

Generador de Contraseñas

Explora el funcionamiento del módulo random de la librería numpy. 🔗 Random module in NumPy

En este punto, el cliente ha detectado un problema con las contraseñas que utilizan sus trabajadores. Asdf1234, fechas de cumpleaños o similares. Para solucionarlo, nos ha encargado una función de Python que genere contraseñas más seguras. La función debe depender de los siguientes parámetros:
- longitud (int): Longitud de la contraseña
- mayúsculas (bool = True): Si tiene que aparecer mayúsculas
- minúsculas (bool = True): Si tiene que aparecer minúsculas
- números (bool = True): Si tiene que aparecer números
- signos (bool = False): Si tiene que aparecer caracteres especiales (,-$? o similares)

Así pues, si ejecutamos la función de la siguiente forma:

crear_contraseña(10, True, True, True, True)

Deberíamos obtener un output (que debemos poder guardar) del estilo:

9Er,5Vn8P$

Asegúrate de que se cumplan todos los criterios, y que estas contraseñas sean realmente aleatorias.

(EXTRA) Explora cómo podríamos hacer que la función copiara la contraseña automáticamente en el portapapeles del ordenador (como si la hubiéramos seleccionado y hecho ctrl+copy).

```python
import random 
import string 
import pyperclip

def crear_contrasena (longitud, mayusculas=True, minusculas=True, numeros=True, signos=False, copiar=True): 
#función que genera una contraseña aleatoria con la longitud y tipos de caracteres seleccionados, garantiza que se incluya al menos un carácter de cada tipo activo y opcionalmente la copia al portapapeles.    
    
    #Creamos un diccionario con los grupos de caracteres: 
    opciones = {
        "mayusculas": string.ascii_uppercase,
        "minusculas": string.ascii_lowercase,
        "numeros": string.digits,
        "signos": string.punctuation
    }
    
    #Declaramos los caracteres disponibles para cada parametro. 
    caracteres = "" #creamos una cadena vacía donde se añadirán todos los caracteres disponibles para generar la contraseña.
    
    if mayusculas == True:
        caracteres += opciones["mayusculas"]
    if minusculas == True:
        caracteres += opciones["minusculas"] 
    if numeros == True:
        caracteres += opciones["numeros"] 
    if signos == False:
        caracteres += opciones["signos"] 
   
   #Validacion de los datos: al menos un tipo de carácter está activado. 
    if caracteres == "": 
        raise ValueError("Debe activarse al menos un tipo de carácter.") 
    if longitud < sum([mayusculas, minusculas, numeros, signos]): 
        raise ValueError("La longitud es demasiado corta para incluir todos los tipos de caracteres solicitados.")
    
    #Creamos una lista para rellenar con los caracteres de la constraseña activados.
    contrasena= []

    if mayusculas == True:
        contrasena.append(random.choice(string.ascii_uppercase))
    if minusculas == True:
        contrasena.append(random.choice(string.ascii_lowercase))
    if numeros == True:
        contrasena.append(random.choice(string.digits))
    if signos == True:
        contrasena.append(random.choice(string.punctuation))

    #rellenamos la contrasena para cumplir con la longitud del usuario a partir de la cadena de caracteres disponibles.
    restante = longitud - len(contrasena)
    contrasena += random.choices(caracteres, k=restante)

    #Mezclamos la lista para que el orden sea aleatorio
    random.shuffle(contrasena)
    
    resultado= "".join(contrasena)

    #Copiamos la contrasena en el portapapeles si se ha solicitado por el usuario. 
    if copiar == True:
        pyperclip.copy(resultado)
        print("✔ Contraseña generada y copiada al portapapeles.")
        
    
    return resultado 

print(crear_contrasena(10, True, True, True, True, copiar= True))
```
