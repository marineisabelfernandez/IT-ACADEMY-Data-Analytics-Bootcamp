- Crea una columna con el nombre completo.

```python
# Creamos una columna con el nombre completo usando el metodo .str.cat()
df["Nombre_completo"] = df["Nom"].str.cat(df["Cognoms"], sep=" " )

# Creamos una columna indicando si la persona nació en España o no.
df["Nacido_espana"]= df["País d'origen"].apply(lambda x: "Si" if x == "Espanya" else "No")

#cambiamos el indice de dataframe por el DNI  
df=df.set_index("DNI")

#Reemplazamos el nombre de las columnas "Día de nacimiento", "Mes de nacimiento" y "Año de nacimiento" por "Día, Mes y Año".
df=df.rename(columns = {'Dia de Naixement':'Día', "Mes de Naixement": "Mes", "Any de Naixement": "Año" })

#Reemplazamos "H" por "Hombre", "D" por "Mujer", "A" por "Otros" y "NC" por un dato que falte (nan/null/na).
df["Gènere"]= df["Gènere"].replace({"D": "Mujer", "H":"Hombre", "A": "Otros", "NC": np.nan})

df
```

## Ejercicio 6 
Crea una nueva columna que sea la fecha de nacimiento en formato Datetime a partir de las columnas día, mes y año. Utilizando esta columna crea una función que dada una fecha, te calcule la edad actual a día de hoy.

Utiliza la función que acabas de crear para generar una columna nueva en DataFrame con la edad actual.

```python
# Creamos la columna “Fecha_nacimiento” uniendo Año, Mes y Día en formato datetime con pd.to_datetime().
from datetime import datetime, date, time, timedelta
df["Fecha_nacimiento"]=pd.to_datetime({"year": df["Año"], "month": df["Mes"],"day": df["Día"]})
df

# Creamos una función que calcula la edad actual a partir de una fecha de nacimiento.
def edad_actual (fecha_nacimiento):
    hoy= datetime.now()
    edad= hoy.year - fecha_nacimiento.year

    if (hoy.month, hoy.day)<(fecha_nacimiento.month, fecha_nacimiento.day): 
        edad -=1 
    return edad 

df["edad_actual"]= df["Fecha_nacimiento"].apply(edad_actual)
df

```

## Ejercicio 1 
Crea una función que tome un dataframe como parámetro de entrada.

La función debe crear (y exportar) un gráfico automáticamente para cada columna del dataframe. Por ejemplo:

- un histograma/boxplot si la variable es numérica
- unas barras de los valores más frecuentes si es categórica
- unas barras de los años más frecuentes si el dato está en formato fecha.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def generar_reporte_grafico(df):
    
    # Validamos que el dataframe no esté vacio
    if df.empty:
            return "Error: El DataFrame no tiene datos."
            
    for columna in df.columns:
         
        try:
            plt.figure() #Creamos una figura vacia para la columna
            
            # Configuramos el grafico para columnas de datos numericos
            if pd.api.types.is_numeric_dtype(df[columna]):
                df[columna].plot.hist(title=f"Distribución: {columna}")
            
            # Configuramos el grafico para columnas de fechas
            elif pd.api.types.is_datetime64_any_dtype(df[columna]):
                df[columna].dt.year.value_counts().sort_index().plot.bar(title=f"Años: {columna}")
            
            #Configuramos el grafico para columnas de datos categoricos    
            else:
                df[columna].value_counts().head(10).plot.bar(title=f"Top 10: {columna}")

            plt.savefig(f"grafico_{columna}.png")
            plt.show() 
            plt.close()
            print(f" Columna {columna} procesada.")

        except Exception as e:
            # Si una columna falla, avisamos y seguimos con la siguiente
            print(f" No se pudo graficar {columna}")
            plt.close()

    return "Gráficos generados y guardados" 

df_muestra = sns.load_dataset('iris')
generar_reporte_grafico(df_muestra) 
```

- Carga el archivo matriz_distancies.xlsx en pandas, de modo que los nombres de filas y los nombres de columnas sean los de las ciudades. Borra "Las Palmas de Gran Canaria" y "Palma" para que podamos realizar el trayecto en coche.

Fuente: Mejoras Rutas

Nos interesa visitar todas las principales ciudades de España recorriendo la mínima distancia posible.

No hace falta que lo hagas de forma óptima, nos interesa que desarrolles una solución razonable utilizando las herramientas que tienes actualmente.

Por ejemplo, una aproximación sencilla (que no óptima) sería yendo siempre a la ciudad más cercana que no hayamos visitado todavía

- Haz una función que dada la matriz de distancias y la ciudad de origen, haga una propuesta de ruta que sea lo más corta posible que puedas, devolviendo una lista con el orden de visita. Da también la distancia total recorrida.

```python
# Cargamos el archivo y le indicamos a pandas que use la primera columna (ciudades) como etiquetas de fila
df_distancias= pd.read_excel(r"C:\Users\marin\OneDrive\Escritorio\DATA ESPECIALIZACION\SPRINT 10\matriu_distancies.xlsx", index_col=0)

# Eliminamos "Las Palmas de Gran Canaria" y "Palma" tanto en filas como en columnas
df_distancias.drop(["Las Palmas de Gran Canaria", "Palma"], axis= 0, inplace=True)
df_distancias.drop(["Las Palmas de Gran Canaria", "Palma"], axis= 1, inplace=True)
df_distancias

def calcular_ruta(df, ciudad_origen):
    ciudades_pendientes = list(df.index)
    ciudades_pendientes.remove(ciudad_origen)
    
    ruta = [ciudad_origen]
    distancia_total = 0
    ciudad_actual = ciudad_origen
    
    for i in range(len(ciudades_pendientes)):
        # 1. Miramos distancias desde donde estamos a las que faltan
        opciones = df.loc[ciudad_actual, ciudades_pendientes] #filtramos filas y columnas
        
        # 2. Elegimos la más cercana
        proxima_ciudad = opciones.idxmin()
        distancia_a_proxima = opciones.min()
        
        # 3. Actualizamos datos
        distancia_total += distancia_a_proxima
        ruta.append(proxima_ciudad)
        
        # 4. Preparamos el siguiente ciclo
        ciudades_pendientes.remove(proxima_ciudad)
        ciudad_actual = proxima_ciudad
        
    return f"itinerario : {ruta}", f"{distancia_total} km recorridos"

Ciudades= df_distancias
Punto_partida= "Zaragoza"
print(f"ruta desde {Punto_partida}")
calcular_ruta(Ciudades,Punto_partida )
```
