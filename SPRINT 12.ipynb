## Nivel 1 - Exploración básica con una API de laboratorio.

## EJERCICO 1 Y 2


1. Consulta la API pública JSONPlaceholder utilizando el método GET para obtener:

- Lista de publicaciones (/posts)

- Lista de usuarios (/users)

- Lista de tareas (/todos)

2. Muestra por pantalla:

- La cantidad total de cada recurso.

- El código de estado de cada petición.

```python
import requests

# URLs de los recursos
url_posts = "https://jsonplaceholder.typicode.com/posts"
url_users = "https://jsonplaceholder.typicode.com/users"
url_todos = "https://jsonplaceholder.typicode.com/todos"

# Realizar las peticiones GET
res_posts = requests.get(url_posts)
res_users = requests.get(url_users)
res_todos = requests.get(url_todos)

# Mostrar resultados por pantalla
print(f"Posts -> Código de estado: {res_posts.status_code} | Total recursos: {len(res_posts.json())}")
print(f"Users -> Código de estado: {res_users.status_code} | Total recursos: {len(res_users.json())}")
print(f"Todos -> Código de estado: {res_todos.status_code} | Total recursos: {len(res_todos.json())}")
```

## EJERCICIO 3
Haz una petición a una publicación inexistente para obtener un error 404 y muestra el código de estado recibido.

```python
# Solicitamos un post con un ID que sepamos que no existe (ej. 9999)
url_inexistente = "https://jsonplaceholder.typicode.com/posts/9999"
res_404 = requests.get(url_inexistente)

print(f"Petición a recurso inexistente -> Código de estado: {res_404.status_code}")
```

## EJERCICIO 4
- Haz una petición POST para crear una nueva publicación ficticia. Incluye el título, el cuerpo del mensaje y un userId.

- Muestra la respuesta JSON.

- Muestra el código de estado.

```python
url_crear = "https://jsonplaceholder.typicode.com/posts"

nuevo_post = {
    "title": "Mi primera publicación de prueba",
    "body": "Este es el cuerpo del mensaje para la tarea de APIs.",
    "userId": 1
}

res_post = requests.post(url_crear, json=nuevo_post)

print(f"Código de estado POST: {res_post.status_code}")
print("Respuesta JSON:")
print(res_post.json())
```

## EJERCICIO 5
- Haz una petición PATCH para modificar parcialmente una publicación existente.

- Muestra la respuesta JSON.

- Muestra el código de estado.

```python
# Modificaremos el post con ID 1
url_patch = "https://jsonplaceholder.typicode.com/posts/1"

datos_modificados = {
    "title": "Título modificado con PATCH"
}

res_patch = requests.patch(url_patch, json=datos_modificados)

print(f"Código de estado PATCH: {res_patch.status_code}")
print("Respuesta JSON:")
print(res_patch.json())
```

## EJERCICIO 6
- Haz una petición DELETE sobre una publicación.

- Muestra la respuesta JSON.

- Muestra el código de estado.

```python
# Eliminaremos el post con ID 1
url_delete = "https://jsonplaceholder.typicode.com/posts/1"

res_delete = requests.delete(url_delete)

print(f"Código de estado DELETE: {res_delete.status_code}")
print("Respuesta JSON (suele estar vacío en JSONPlaceholder):")
print(res_delete.json())
```

### Documentación de la API elegida: Open-Meteo
Hemos elegido la API de Open-Meteo para consultar datos meteorológicos actuales e históricos.

**Endpoints disponibles:**
1. Forecast (Predicción): `https://api.open-meteo.com/v1/forecast`
2. Historical Weather (Histórico): `https://archive-api.open-meteo.com/v1/archive`

**Filtros y parámetros opcionales utilizados:**
* `latitude` y `longitude`: Coordenadas geográficas (obligatorio, usaremos las de Barcelona: 41.3887, 2.159).
* `hourly`: Para solicitar variables específicas por horas (ej. `temperature_2m`, `relative_humidity_2m`).
* `timezone`: Para ajustar la zona horaria (ej. `Europe/Berlin`).

**Formato de respuesta:** JSON.

## EJERCICIOS 3 Y 4
3. Haz una petición GET sencilla:

- Muestra el código de estado de la respuesta.

- Imprime de forma clara algunos de los campos de la respuesta JSON.

4. Convierte los datos a un DataFrame de pandas:

- Muestra las primeras filas del DataFrame.


```python
import pandas as pd 

# URL del endpoint con los parámetros para Barcelona
url_weather = "https://api.open-meteo.com/v1/forecast"
params = {
    "latitude": 41.3887,
    "longitude": 2.159,
    "hourly": "temperature_2m,relative_humidity_2m",
    "timezone": "Europe/Berlin"
}

res_weather = requests.get(url_weather, params=params)
print(f"Código de estado de la respuesta: {res_weather.status_code}\n")

# Extraer el JSON
data = res_weather.json()

# Imprimir de forma clara algunos campos clave del JSON
print("--- Campos clave de la respuesta ---")
print(f"Latitud: {data['latitude']}")
print(f"Longitud: {data['longitude']}")
print(f"Zona horaria: {data['timezone']}")
print(f"Unidades de medida: {data['hourly_units']}\n")

# Convertir la sección 'hourly' de los datos a un DataFrame de pandas
df_weather = pd.DataFrame(data['hourly'])

# Mostrar las primeras filas del DataFrame
print("--- Primeras filas del DataFrame ---")
df_weather.head()
```

# NIVEL 3 - API de Open Data Barcelona

## EJERCICIOS 1 A 5 
1. Utiliza la API de Open Data BCN para:
Buscar un dataset de tu interés mediante package_search y package_show.

2. De los resultados obtenidos, selecciona uno que tenga recursos disponibles en CSV o JSON.

3. Con el resource_id del recurso seleccionado, realiza una consulta mediante datastore_search para:
Recuperar al menos 100 registros.

4. Convierte los resultados a un DataFrame.

5. Guarda el DataFrame en un archivo .csv.

```python
import requests
import pandas as pd

# 1. Buscar un dataset utilizando package_search
url_search = "https://opendata-ajuntament.barcelona.cat/data/api/3/action/package_search"
params_search = {
    "q": "accidents", # Buscaremos datasets relacionados con accidentes
    "rows": 5
}

res_search = requests.get(url_search, params=params_search)
search_data = res_search.json()

# Seleccionamos el primer paquete/dataset del resultado
package = search_data['result']['results'][0]
package_id = package['id']
print(f"Dataset seleccionado: {package['title']} (ID: {package_id})")

# 2. Explorar los recursos del dataset elegido para encontrar el formato adecuado (CSV o JSON)
# Usamos package_show para ver al detalle sus recursos
url_show = "https://opendata-ajuntament.barcelona.cat/data/api/3/action/package_show"
res_show = requests.get(url_show, params={"id": package_id})
show_data = res_show.json()

resource_id = None
resource_name = ""

for resource in show_data['result']['resources']:
    # Buscamos preferiblemente un recurso que esté integrado en el Datastore (API consultable)
    if resource['datastore_active'] == True:
        resource_id = resource['id']
        resource_name = resource['name']
        break

# Si no encontrara uno con datastore activo, agarramos el primero que sea JSON o CSV
if not resource_id:
    for resource in show_data['result']['resources']:
        if resource['format'].lower() in ['json', 'csv']:
            resource_id = resource['id']
            resource_name = resource['name']
            break

print(f"Recurso seleccionado: {resource_name} (Resource ID: {resource_id})")

# 3. Realizar una consulta mediante datastore_search para recuperar al menos 100 registros
url_datastore = "https://opendata-ajuntament.barcelona.cat/data/api/3/action/datastore_search"
params_datastore = {
    "resource_id": resource_id,
    "limit": 100 # Pedimos explícitamente 100 registros
}

res_datastore = requests.get(url_datastore, params=params_datastore)
datastore_data = res_datastore.json()

# Los registros reales están dentro de 'records'
records = datastore_data['result']['records']
print(f"Cantidad de registros recuperados: {len(records)}")

# 4. Convertir los resultados a un DataFrame de pandas
df_bcn = pd.DataFrame(records)
print("\n--- Primeras filas del DataFrame de Barcelona ---")
print(df_bcn.head())

# 5. Guardar el DataFrame en un fichero .csv
nombre_archivo = "accidentes_bcn.csv"
df_bcn.to_csv(nombre_archivo, index=False, encoding='utf-8')
print(f"\n¡DataFrame guardado con éxito en '{nombre_archivo}'!")
```
